```python
import pandas as pd
df_dev = pd.read_csv("data/development.csv")
df_eva = pd.read_csv("data/evaluation.csv")
df_merged = pd.concat([df_eva,df_dev],sort=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation_matrices = {}
for i in range(18):
    columns = [f'pmax[{i}]', f'negpmax[{i}]', f'area[{i}]', f'tmax[{i}]', f'rms[{i}]','x','y']
    df_subset = df_dev[columns]
    correlation_matrix = df_subset.corr("spearman")
    correlation_matrices[f'i={i}'] = correlation_matrix
for key, matrix in correlation_matrices.items():
    plt.figure()
    sns.heatmap(matrix, annot=True, cmap="Blues", fmt=".2f")
    plt.show()
```

```python
columns = []
for i in [0,7,12,15,16,17]:
    columns.append(f"pmax[{i}]")
columns = columns + ["x","y"]
df_subset = df_dev[columns]
correlation_matrix = df_subset.corr("spearman")
plt.figure()
plt.title("Correlation matrix pmax", fontsize=15)
sns.heatmap(correlation_matrix, cmap="coolwarm", fmt=".2f", vmin = -1, vmax = 1)
#plt.show()
plt.savefig("correlation_excluded_pmax.pdf", bbox_inches="tight")

columns = []
for i in [1,2,3,4,5,6,8,9,10,11,13,14]:
    columns.append(f"pmax[{i}]")
columns = columns + ["x","y"]
df_subset = df_dev[columns]
correlation_matrix = df_subset.corr("spearman")
plt.figure(figsize=(16,16))
plt.title("Correlation matrix pmax", fontsize=15)
sns.heatmap(correlation_matrix, cmap="coolwarm", fmt=".2f", vmin = -1, vmax = 1)
#plt.show()
plt.savefig("correlation_included_pmax.pdf", bbox_inches="tight")
```

```python
columns = []
for i in [0,7,12,15,16,17]:
    columns.append(f"negpmax[{i}]")
columns = columns + ["x","y"]
df_subset = df_dev[columns]
correlation_matrix = df_subset.corr("spearman")
plt.figure()
plt.title("Correlation matrix negpmax", fontsize=15)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin = -1, vmax = 1)
#plt.show()
plt.savefig("correlation_excluded_negpmax.pdf", bbox_inches="tight")

columns = []
for i in [1,2,3,4,5,6,8,9,10,11,13,14]:
    columns.append(f"negpmax[{i}]")
columns = columns + ["x","y"]
df_subset = df_dev[columns]
correlation_matrix = df_subset.corr("spearman")
plt.figure(figsize=(16,16))
plt.title("Correlation matrix negpmax", fontsize=15)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin = -1, vmax = 1)
#plt.show()
plt.savefig("correlation_included_negpmax.pdf", bbox_inches="tight")
```

```python
pmax_columns = [f'pmax[{i}]' for i in (1,2,3,4,5,6,8,9,10,11,13,14)]
plt.figure(figsize=(12, 8))
sns.boxplot(df_merged[pmax_columns], width=0.7)
plt.title('Boxplot di pmax')
plt.xlabel('Numero dell\'pmax')
plt.ylabel('Valore dell\'pmax')
plt.savefig("boxplot_pmax.pdf", bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([100, 200, 300, 400, 500])
x2_points = np.array([100, 200, 300, 400,500])
y_points = [4.229853625474316, 4.2068101828442925, 4.194656601360927, 4.190981976202009, 4.188431461889038]
y2_points = [4.662433123110209, 4.637941994296972, 4.63182956268614, 4.627102575378807,4.619731456274208]

plt.rcdefaults()

plt.xlabel("n_estimators")
plt.ylabel("euclidean distance (µm)")
plt.title("Comparison between n_estimators and euclidean distance with fixed max_depth")
bar_width = 15
plt.bar(x_points - bar_width/2, y_points, width=bar_width, label="Random Forest")
plt.bar(x2_points + bar_width/2, y2_points, width=bar_width, color="red", label="Extra Trees")

plt.legend()
plt.grid()
plt.savefig("bar_graph.pdf", bbox_inches="tight")
plt.show()
```
